# **CHARTS Design**

### **Monthly Sales Trend**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str) 

monthly_revenue = df.groupby('YearMonth')['Price (INR)'].sum().reset_index()

plt.figure()
plt.plot(monthly_revenue['YearMonth'], monthly_revenue['Price (INR)'])
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.tight_layout()
plt.show()
```

### **Daily Sales Trend**

```python
df['DayName'] = pd.to_datetime(df['Order Date']).dt.day_name()

daily_revenue = (
    df.groupby('DayName')['Price (INR)']
    .sum()
    .reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
)

plt.figure(figsize=(10,5))
plt.bar(daily_revenue.index, daily_revenue.values)
plt.title('Daily Revenue Trend (Mon-Sun)')
plt.xlabel("Day")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=30)
plt.show()
```

### **Total Sales by Food Type (Veg vs Non-Veg)**

```python
non_veg_keywords = [
    'chicken', 'egg', 'fish', 'mutton',
    'prawn', 'biryani', 'kabab', 'kebab',
    "non-veg", 'non veg'
]

df['Food Category'] = np.where(
    df['Dish Name'].str.lower().str.contains('|'.join(non_veg_keywords), na=False),
    "Non-Veg",
    'Veg'
)
```

```python
food_revenue = (
    df.groupby('Food Category')['Price (INR)']
    .sum()
    .reset_index()
)
```

```python
fig = px.pie(
    food_revenue,
    values= "Price (INR)",
    names = "Food Category",
    hole=0.5,
    title='Revenue Contribution: Veg vs Non-Veg',
)

fig.update_traces(
    textinfo='percent+label',
    pull=[0.05, 0]
)

fig.update_layout(
    height=500,
    margin=dict(t=60, b=40, l=40, r=40)
)

fig.show()
```

### **Total Sales by State**

```python
# Strip spaces and replace spaces with underscores
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(df.columns.tolist())
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'  # Open chart in browser

# Aggregate revenue by state
state_revenue = df.groupby("State", as_index=False)['Price_INR'].sum()
state_revenue = state_revenue.sort_values('Price_INR', ascending=False)

# Horizontal bar chart
fig = px.bar(
    state_revenue,
    x='Price_INR',
    y='State',
    orientation='h',
    title='Revenue by State (INR)'
)

fig.update_layout(height=600, yaxis=dict(autorange='reversed'))
fig.show()
```

### **Quarterly Performance Summary**

```python
import pandas as pd
import plotly.express as px
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Convert order date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Add quarter column
df['Quarter'] = df['Order_Date'].dt.to_period('Q').astype(str)

# Aggregate data by quarter
quarterly_summary = (
    df.groupby('Quarter', as_index=False)
    .agg(
        Total_Sales=('Price_INR', 'sum'),
        Avg_Rating=('Rating', 'mean'),
        Total_Orders=('Order_Date', 'count')
    )
    .sort_values("Quarter")
)

# Round values
quarterly_summary['Total_Sales'] = quarterly_summary['Total_Sales'].round(0)
quarterly_summary['Avg_Rating'] = quarterly_summary['Avg_Rating'].round(2)

# Display the result
print(quarterly_summary)
```

### **Top 5 Cities by Sales**

```python
import pandas as pd
import plotly.express as px

# Clean column names to avoid KeyErrors
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Compute top 5 cities by total sales
top_5_cities = (
    df.groupby('City')['Price_INR']  # use cleaned column name
    .sum()
    .nlargest(5)
    .sort_values()
    .reset_index()
)

# Create horizontal bar chart
fig = px.bar(
    top_5_cities,
    x='Price_INR',  # cleaned column name
    y='City',
    orientation='h',
    title='Top 5 Cities by Sales (INR)',
    color_discrete_sequence=['red']
)

fig.update_layout(height=400)  # optional: adjust chart height
fig.show()
```

### **Weekly Trend Analysis**

```python
import pandas as pd
import plotly.express as px

# 1 Clean column names to avoid KeyErrors
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# 2 Convert order date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# 3 Extract week number and year
df['Year_Week'] = df['Order_Date'].dt.strftime('%Y-%U')  # %U gives week number

# 4 Aggregate weekly sales
weekly_sales = (
    df.groupby('Year_Week', as_index=False)
    .agg(
        Total_Sales=('Price_INR', 'sum'),
        Total_Orders=('Order_Date', 'count'),
        Avg_Rating=('Rating', 'mean')
    )
)

# Optional: Round numeric columns
weekly_sales['Total_Sales'] = weekly_sales['Total_Sales'].round(0)
weekly_sales['Avg_Rating'] = weekly_sales['Avg_Rating'].round(2)

# 5 Plot weekly sales trend
fig = px.line(
    weekly_sales,
    x='Year_Week',
    y='Total_Sales',
    title='Weekly Sales Trend',
    markers=True
)
fig.update_layout(xaxis_title='Week', yaxis_title='Total Sales (INR)', xaxis_tickangle=-45)
fig.show()
```
